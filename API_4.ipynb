```python
What is REST API?

REST (Representational State Transfer) is a set of rules and principles for designing APIs.

REST defines:

How URLs should look

How HTTP methods should be used
GET ‚Üí Read data

POST ‚Üí Create data

PUT ‚Üí Update data

DELETE ‚Üí Delete data    

How data should be exchanged (JSON/XML)
```

```python
FastAPI 

FastAPI is a Python web framework specifically designed for building APIs (Application Programming Interfaces).

It helps you easily create HTTP endpoints, handle requests, validate data, and return responses in JSON.
It also provide Tools to handle query parameters, path parameters, and request bodies

It also provides automatic documentation (Swagger UI and ReDoc) without extra effort.
```

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Hello FastAPI"}
```

```python
1.from fastapi import FastAPI
This line imports the FastAPI class from the fastapi library.
FastAPI is a framework that helps you easily create web APIs in Python.
2.app = FastAPI()
Here, we are creating an instance of the FastAPI class and storing it in the variable app
3.@app.get("/")
This is a decorator. A decorator ‚Äúwraps‚Äù a function to add extra functionality.

@app.get("/") means:

‚ÄúWhenever someone sends a GET request to the root path / of this API, run the following function.‚Äù
/ ‚Üí The root endpoint, i.e., the top-level URL of your API.
GET ‚Üí The HTTP method used to fetch data from the server.
4.def read_root():
This defines a Python function named read_root.
This function is executed when someone visits / with a GET request.
The function contains the logic for what the server should do when this endpoint is called.
5.return {"message": "Hello FastAPI"}
The function returns a Python dictionary.
‚úÖ Response is JSON by default
FastAPI automatically converts this dictionary into JSON, which is the standard format for APIs.
```

```python
‚úÖ The root endpoint is just the main entrance to your API.represented by /
eg.http://127.0.0.1:8000/
```

```python
Endpoint	                HTTP Method	Description
/	                         GET	Root endpoint, basic info
/users	                        GET	Get list of users
/users	POST	               Create a new user
/users/1	                    GET	Get user with ID 1
```

```python
#Perfect! Let‚Äôs add a new endpoint /hello that accepts a name as a query parameter and returns a greeting. 
#Here‚Äôs how you can update your main.py:

from fastapi import FastAPI

app = FastAPI()

# Root endpoint 
@app.get("/")
def read_root():
    return {"message": "Hello World"}

# New endpoint: /hello?name=Renu
@app.get("/hello")
def say_hello(name: str = "Guest"):
    return {"message": f"Hello, {name}!"}
```

```python
1Ô∏è‚É£ How it works

@app.get("/hello") ‚Üí defines a new GET endpoint /hello.

name: str = "Guest" ‚Üí name is a query parameter; 
It means that the parameter name should be a string (str).
Type  help FastAPI automatically validate input.
If someone sends a number instead of a string in the query, FastAPI can raise an error or handle it correctly.
2Ô∏è‚É£ = "Guest"

This is a default value.
If the user does not provide the name in the query parameter, it will automatically use "Guest".
For example:

URL	Result
/hello?name=Renu	{"message": "Hello, Renu!"}
/hello	{"message": "Hello, Guest!"}
return {"message": f"Hello, {name}!"} ‚Üí returns a JSON response with your greeting.
```

```python
1.http://127.0.0.1:8000/
output :{"message":"Hello FastAPI"}
2..http://127.0.0.1:8000/hello
output:{"message":"Hello, Guest!"}
3.http://127.0.0.1:8000/hello?name=Renu
output:{"message":"Hello, Renu!"}
```
